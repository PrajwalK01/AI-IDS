```python
# Cell 1: Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cell 2: Load dataset (example NSL-KDD)
df = pd.read_csv("../data/NSL-KDD.csv")  # adjust path if needed
df.head()

# Cell 3: Basic info
print(df.shape)
print(df.info())

# Cell 4: Class distribution
sns.countplot(x="label", data=df)
plt.title("Attack vs Normal Distribution")
plt.show()

# Cell 5: Feature correlations
corr = df.corr(numeric_only=True)
plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()

```
